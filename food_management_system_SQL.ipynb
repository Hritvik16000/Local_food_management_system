```python
import pandas as pd
from sqlalchemy import create_engine, text

DATABASE_URL = "postgresql://hritvikdadhich@localhost:5432/food_wastage_db"
engine = create_engine(DATABASE_URL)

print("--- Question 1a: Number of Food Providers in Each City ---")
query_providers_per_city = """
SELECT
    city,
    COUNT(provider_id) AS NumberOfProviders
FROM
    providers
GROUP BY
    city
ORDER BY
    NumberOfProviders DESC;
"""

try:
    df_providers_per_city = pd.read_sql(query_providers_per_city, engine)
    print(df_providers_per_city)
    print("\n" + "="*80 + "\n")
except Exception as e:
    print(f"Error executing query for providers per city: {e}")

print("--- Question 1b: Number of Receivers in Each City ---")
query_receivers_per_city = """
SELECT
    city,
    COUNT(receiver_id) AS NumberOfReceivers
FROM
    receivers
GROUP BY
    city
ORDER BY
    NumberOfReceivers DESC;
"""

try:
    df_receivers_per_city = pd.read_sql(query_receivers_per_city, engine)
    print(df_receivers_per_city)
except Exception as e:
    print(f"Error executing query for receivers per city: {e}")
```

```python
import pandas as pd
from sqlalchemy import create_engine, text


DATABASE_URL = "postgresql://hritvikdadhich@localhost:5432/food_wastage_db"
engine = create_engine(DATABASE_URL)

print("--- Question 9: Which provider has had the highest number of successful food claims? ---")
query_provider_successful_claims = """
SELECT
    p.name AS ProviderName, -- Name of the provider
    COUNT(c.claim_id) AS NumberOfSuccessfulClaims -- Count of claims with 'Completed' status
FROM
    claims c
JOIN
    food_listings fl ON c.food_id = fl.food_id -- Join to link claims to food listings
JOIN
    providers p ON fl.provider_id = p.provider_id -- Join to link food listings to providers
WHERE
    c.status = 'Completed' -- Filter for successful claims
GROUP BY
    p.name -- Group by provider name
ORDER BY
    NumberOfSuccessfulClaims DESC -- Order to find the provider with the most successful claims
LIMIT 1; -- Limit to 1 to show only the top provider
"""

try:
    df_provider_successful_claims = pd.read_sql(query_provider_successful_claims, engine)
    print(df_provider_successful_claims)
    print("\n" + "="*80 + "\n")

except Exception as e:
    print(f"Error executing query for provider with successful claims: {e}")
```

```python
import pandas as pd
from sqlalchemy import create_engine, text


DATABASE_URL = "postgresql://hritvikdadhich@localhost:5432/food_wastage_db"
engine = create_engine(DATABASE_URL)

print("--- Question 10: What percentage of food claims are completed vs. pending vs. canceled? ---")
query_claim_status_percentage = """
SELECT
    status, -- Status of the claim (Completed, Pending, Cancelled)
    COUNT(claim_id) AS NumberOfClaims, -- Count of claims for each status
    (COUNT(claim_id) * 100.0 / (SELECT COUNT(*) FROM claims)) AS Percentage -- Calculate percentage
FROM
    claims
GROUP BY
    status
ORDER BY
    Percentage DESC;
"""

try:
    df_claim_status_percentage = pd.read_sql(query_claim_status_percentage, engine)
    print(df_claim_status_percentage)
    print("\n" + "="*80 + "\n")

except Exception as e:
    print(f"Error executing query for claim status percentage: {e}")
```

```python
import pandas as pd
from sqlalchemy import create_engine, text


DATABASE_URL = "postgresql://hritvikdadhich@localhost:5432/food_wastage_db"
engine = create_engine(DATABASE_URL)

print("--- Question 11: What is the average quantity of food claimed per receiver? ---")
query_avg_quantity_claimed_per_receiver = """
SELECT
    AVG(TotalQuantityClaimed) AS AverageQuantityClaimedPerReceiver
FROM (
    SELECT
        c.receiver_id, -- Receiver ID
        SUM(fl.quantity) AS TotalQuantityClaimed -- Sum of quantity claimed by each receiver
    FROM
        claims c
    JOIN
        food_listings fl ON c.food_id = fl.food_id -- Join to get food quantity
    GROUP BY
        c.receiver_id -- Group by receiver to sum their total claims
) AS ReceiverClaims;
"""

try:
    df_avg_claimed_per_receiver = pd.read_sql(query_avg_quantity_claimed_per_receiver, engine)
    print(df_avg_claimed_per_receiver)
    print("\n" + "="*80 + "\n")

except Exception as e:
    print(f"Error executing query for average quantity claimed per receiver: {e}")
```
